```python
import pandas as pd

pd.options.display.float_format = '{:.2f}'.format
pd.set_option('display.max_columns', 100)
```

```python
df_cand = pd.read_csv('datasets/consulta_cand_2018_BRASIL.csv', sep=';', encoding = 'ISO-8859-1')
df_cand.head(3)
```

```python
# Tratando erros pois há candidado com data de nascimento no ano de 1193
df_cand['DT_NASCIMENTO'] = pd.to_datetime(df_cand['DT_NASCIMENTO'], errors='coerce').dt.strftime('%Y-%m-%d')
```

```python
print(df_cand.shape)
print(df_cand['SQ_CANDIDATO'].nunique())
```

```python
df_bens = pd.read_csv('datasets/bem_candidato_2018_BRASIL.csv', sep=';', encoding = 'ISO-8859-1', decimal=',')
df_bens.head(3)
```

```python
df_bens2 = df_bens[['SQ_CANDIDATO','CD_TIPO_BEM_CANDIDATO','DS_TIPO_BEM_CANDIDATO','DS_BEM_CANDIDATO','VR_BEM_CANDIDATO']]
df_bens2.head(3)
```

```python
df_bens2.dtypes
```

```python
df_bens2.shape
```

```python
df_bens_por_cand = df_bens2.groupby('SQ_CANDIDATO').apply(lambda x: x.to_dict(orient='records')).to_frame().reset_index()
df_bens_por_cand.columns = ['SQ_CANDIDATO','BENS']
```

```python
print(df_bens_por_cand.dtypes)
df_bens_por_cand['BENS'][0]
```

```python
df_bens_merged = pd.merge(left=df_cand, right=df_bens_por_cand, on='SQ_CANDIDATO', how='left')
df_bens_merged.head(5)
```

```python
df_bens_merged.shape
```

```python
df_bens_merged.to_json('datasets/export.json', orient='records', lines=True, date_format='iso')
```
